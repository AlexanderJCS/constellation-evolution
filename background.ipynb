## Galactic Gravitational Field

This [Stack Overflow answer](https://physics.stackexchange.com/a/161615) is very helpful for summarizing the different models of galactic gravitational fields.

There are three main sources of gravitational potential: the disk, the bulge, and the dark matter halo.

![Image of the disk, bulge, and dark matter halo](https://cdn.sci.news/images/2019/12/image_7881-Milky-Way-Disks.jpg)

Taking from the StackOverflow response:

The disk has two common models:
* Kuzmin model:
$$\Phi(r,z)=-\frac{GM}{\sqrt{r^2+(a+|z|)^2}}$$
* Miyamoto-Nagai model:
$$\Phi(r,z)=-\frac{GM}{\sqrt{r^2+(a+\sqrt{z^2+b^2})^2}}$$
where $a$ and $b$ are scale lengths.

The bulge can be represented with the following spherically symmetric models:
 - Plummer model:
$$\Phi(r)=−\frac{GM}{\sqrt{r^2+a^2}}$$
 - Jaffe model:
$$\Phi(r)=\frac{GM}{a}\ln\left(\frac{r}{r+a}\right)$$
where $a$ also is a scale length and not necessarily the same as those for the disk.

The dark halo takes a spherical form,
$$\Phi(r)=\frac12V_h^2\ln\left(r^2+a^2\right)$$
where $V_h$ is the radial velocity of the galaxy at far distances ($\sim200$ km/s) and $a$ another scale length that isn't necessarily the same as above.

I'm curious how these models differ from each other, so I'm going to plot them together

```python
G = 1.0
M = 1.0
a = 1.0
b = 1.0
```

```python
import numpy as np

def kuzmin(r, z=0):
    return -G * M / np.sqrt(r**2 + (a + abs(z))**2)


def miyamoto_nagai(r, z=0):
    return -G * M / np.sqrt(r**2 + (a + np.sqrt(z**2 + b**2))**2)
```

```python
import matplotlib.pyplot as plt
plt.style.use("default")
%config InlineBackend.figure_format = "retina"
```

```python
def graph(f, lower=0, upper=10, label=None):
    points = np.linspace(lower, upper, 1000)
    plt.plot(points, f(points), label=label)
```

```python
graph(kuzmin, label="Kuzmin")
graph(miyamoto_nagai, label="Miyamoto-Nagai")
plt.legend()
plt.show()
```
